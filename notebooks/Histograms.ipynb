```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('dark_background')
```

```python
benin = pd.read_csv("../data/benin_checked.csv", parse_dates=['Timestamp'])
```

```python
columns_to_plot = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS']


for col in columns_to_plot:
    benin[col] = pd.to_numeric(benin[col], errors='coerce')

```

```python
columns_to_plot = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS']
for col in columns_to_plot:
    print(f"{col}: {benin[col].dtype} — {benin[col].isna().sum()} missing")

```

```python
columns_to_plot = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS']

for col in columns_to_plot:
    plt.figure(figsize=(8, 4))
    sns.histplot(benin[col].dropna(), bins=100, kde=True)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```
